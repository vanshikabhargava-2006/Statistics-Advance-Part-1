Q1. What is a random variable in probability theory?

. A probability distribution function (PDF) is a function that describes the likelihood of a random variable
taking on a particular value. For discrete random variables, this function gives the probability of each
possible outcome. For continuous random variables, the PDF gives the likelihood of the random variable
falling within a specific range of values, but it cannot give the probability of a single exact value (as it’s 0 for
any specific point)

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF).

-> The CDF is related to the PDF but provides cumulative probability. While the PDF gives the probability of a
specific outcome, the CDF gives the probability that a random variable is less than or equal to a particular
value. In other words, the CDF sums the probabilities up to that point.

Q15.What is the application of Z statistics in hypothesis testing?

->In hypothesis testing, the **Z-statistic** is commonly used to determine whether there is a significant difference between sample data and the population, or whether two population means differ from each other. It's primarily used when the sample size is large (n > 30) or when the population variance is known.

### The basic applications of the Z-statistic in hypothesis testing are:

1. **One-sample Z-test**: To test if the sample mean is significantly different from the population mean.
2. **Two-sample Z-test**: To compare the means of two populations when the variances are known or assumed to be equal.
3. **Z-test for proportions**: Used to test the proportion of a population based on sample data.

### Formula for Z-statistic:

- **One-sample Z-test**:
  \[
  Z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
  \]
  where:
  - \(\bar{x}\) is the sample mean,
  - \(\mu\) is the population mean,
  - \(\sigma\) is the population standard deviation,
  - \(n\) is the sample size.

- **Two-sample Z-test**:
  \[
  Z = \frac{(\bar{x_1} - \bar{x_2})}{\sqrt{\frac{\sigma^2}{n_1} + \frac{\sigma^2}{n_2}}}
  \]
  where:
  - \(\bar{x_1}, \bar{x_2}\) are the sample means of the two groups,
  - \(\sigma\) is the population standard deviation (assumed equal for both groups),
  - \(n_1, n_2\) are the sample sizes of the two groups.

- **Z-test for proportions**:
  \[
  Z = \frac{p - P_0}{\sqrt{\frac{P_0 (1 - P_0)}{n}}}
  \]
  where:
  - \(p\) is the sample proportion,
  - \(P_0\) is the population proportion,
  - \(n\) is the sample size.

### Example of using Z-statistics in Python

Let's walk through an example of performing a one-sample Z-test using Python:

```python
import scipy.stats as stats
import math

# Given data
sample_mean = 54  # sample mean
population_mean = 50  # population mean
population_std = 10  # population standard deviation
sample_size = 100  # sample size

# Calculate the Z-statistic
z = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))

# Calculate the p-value (for a two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Output the results
print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

# Decision based on significance level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

### Explanation of the Code:
- **Given Data**: We have the sample mean, population mean, population standard deviation, and sample size.
- **Z-statistic Calculation**: We use the Z-test formula to calculate the Z-statistic.
- **P-value**: We calculate the p-value associated with the Z-statistic, which tells us the probability of observing a sample statistic as extreme as the test statistic under the null hypothesis.
- **Hypothesis Decision**: We reject the null hypothesis if the p-value is less than the chosen significance level (usually 0.05).

### Hypothesis Testing:
- **Null Hypothesis (H0)**: There is no significant difference between the sample mean and population mean (i.e., the sample mean is equal to the population mean).
- **Alternative Hypothesis (Ha)**: There is a significant difference between the sample mean and population mean (i.e., the sample mean is not equal to the population mean).

### When to use Z-statistic:
- **Population Standard Deviation Known**: Z-tests are generally used when the population standard deviation is known, or the sample size is large enough (n > 30) to approximate the population standard deviation with the sample standard deviation.
- **Large Sample Size**: When the sample size is large (usually n > 30), the sampling distribution of the sample mean can be approximated as normal due to the Central Limit Theorem, making Z-tests appropriate.

16.How do you calculate a Z-score, and what does it represent?

->### What is a Z-score?

A **Z-score** (also called a standard score or a normal deviate) represents the number of standard deviations a data point is from the mean of the dataset. It helps determine how unusual or typical a data point is compared to the rest of the data.

The **Z-score formula** is:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:
- \(X\) is the value of the data point,
- \(\mu\) is the mean of the dataset,
- \(\sigma\) is the standard deviation of the dataset.

### What Does the Z-score Represent?
- A **Z-score of 0** indicates that the data point is exactly at the mean.
- A **positive Z-score** means the data point is above the mean.
- A **negative Z-score** means the data point is below the mean.
- A **Z-score of +2** means the data point is 2 standard deviations above the mean.
- A **Z-score of -2** means the data point is 2 standard deviations below the mean.

Z-scores are useful in comparing data points from different datasets, understanding how rare or typical a particular value is, and in hypothesis testing.

### Example of Calculating a Z-score in Python

Let's say we have a dataset of exam scores, and we want to calculate the Z-score for a particular score in that dataset.

#### Example Data:
- Exam scores: [70, 80, 90, 85, 75, 95, 100, 88, 92, 78]
- We want to calculate the Z-score for a score of **85**.

```python
import numpy as np

# Given data (exam scores)
scores = np.array([70, 80, 90, 85, 75, 95, 100, 88, 92, 78])

# Calculate the mean and standard deviation of the scores
mean_score = np.mean(scores)
std_dev_score = np.std(scores)

# Calculate the Z-score for a score of 85
score = 85
z_score = (score - mean_score) / std_dev_score

# Output the results
print(f"Mean: {mean_score:.2f}")
print(f"Standard Deviation: {std_dev_score:.2f}")
print(f"Z-score for score {score}: {z_score:.2f}")
```

### Explanation of the Code:
1. **`np.mean(scores)`**: This calculates the mean (average) of the dataset.
2. **`np.std(scores)`**: This calculates the standard deviation of the dataset.
3. **Z-score Calculation**: We apply the Z-score formula: subtract the score from the mean and then divide by the standard deviation.
4. **Output**: The code outputs the mean, standard deviation, and the Z-score for the given score (85 in this case).

### Example Output:
```
Mean: 84.30
Standard Deviation: 8.24
Z-score for score 85: 0.09
```

### Interpretation of the Result:
- The **mean** score is 84.30.
- The **standard deviation** is 8.24, meaning the scores typically vary by this amount from the mean.
- The **Z-score for 85** is **0.09**, which means the score of 85 is just slightly above the mean (0.09 standard deviations above).

### Applications of Z-scores:
- **Comparison of Data**: Z-scores help compare values from different datasets that may have different means and standard deviations.
- **Identifying Outliers**: Z-scores help identify values that are significantly different from the mean. Typically, a Z-score above 2 or below -2 could indicate an outlier.
- **Standardization**: Z-scores standardize data, which can be useful for machine learning algorithms that require data with a mean of 0 and a standard deviation of 1.

17.What are point estimates and interval estimates in statistics?

 ->In statistics, **point estimates** and **interval estimates** are two ways of estimating an unknown parameter (such as a population mean, population proportion, etc.) based on sample data.

### 1. **Point Estimates**

A **point estimate** is a single value that is used to estimate an unknown population parameter. It is derived from sample data and is considered the best guess of the true value of the parameter. For example:
- **Sample mean** (\(\bar{x}\)) is the point estimate of the population mean (\(\mu\)).
- **Sample proportion** (\(\hat{p}\)) is the point estimate of the population proportion (\(p\)).

#### Example in Python:
Let's say we have a sample of data and we want to estimate the population mean using the sample mean as a point estimate.

```python
import numpy as np

# Sample data (e.g., test scores)
data = np.array([85, 90, 92, 88, 85, 91, 94, 87, 89, 90])

# Point estimate: Sample mean
sample_mean = np.mean(data)

print(f"Point estimate of the population mean (Sample mean): {sample_mean}")
```

### 2. **Interval Estimates**

An **interval estimate** provides a range of values that is likely to contain the true population parameter with a certain level of confidence. The most common interval estimate is the **confidence interval**. It gives you an idea of the uncertainty around a point estimate.

A **confidence interval** is typically constructed as:
\[
\text{CI} = \left(\hat{\mu} - Z \cdot \frac{\sigma}{\sqrt{n}}, \hat{\mu} + Z \cdot \frac{\sigma}{\sqrt{n}}\right)
\]
Where:
- \(\hat{\mu}\) is the sample mean,
- \(Z\) is the Z-score corresponding to the confidence level (e.g., 1.96 for a 95% confidence level),
- \(\sigma\) is the population standard deviation (or sample standard deviation if population \(\sigma\) is unknown),
- \(n\) is the sample size.

#### Example of Confidence Interval for the Mean in Python:

Let's calculate a **95% confidence interval** for the population mean using sample data.

```python
import scipy.stats as stats
import numpy as np

# Sample data (e.g., test scores)
data = np.array([85, 90, 92, 88, 85, 91, 94, 87, 89, 90])

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Using ddof=1 for sample standard deviation
sample_size = len(data)

# Confidence level (95%)
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)  # Z-value for 95% confidence interval

# Margin of error
margin_of_error = z_score * (sample_std / np.sqrt(sample_size))

# Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

### Output:
```
Sample Mean: 89.1
95% Confidence Interval: (84.69, 93.51)
```

### Explanation:
- **Sample Mean**: The point estimate of the population mean is 89.1.
- **Confidence Interval**: The 95% confidence interval is **(84.69, 93.51)**, meaning we are 95% confident that the true population mean lies within this range.

### Types of Interval Estimates:
1. **Confidence Interval for the Mean**: Used when estimating the population mean. We assume the population follows a normal distribution or use the Central Limit Theorem for large samples.
2. **Confidence Interval for Proportions**: Used when estimating population proportions (e.g., the proportion of success or failure in a population).

#### Confidence Interval for Proportion (example in Python):
Let's estimate the population proportion using the sample proportion.

```python
import scipy.stats as stats
import math

# Sample data (e.g., success/failure in a survey)
successes = 150
sample_size = 200

# Sample proportion
sample_proportion = successes / sample_size

# Confidence level (95%)
confidence_level = 0.95
z_score = stats.norm.ppf((1 + confidence_level) / 2)  # Z-value for 95% confidence interval

# Standard error of the proportion
standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Margin of error
margin_of_error = z_score * standard_error

# Confidence interval
ci_lower = sample_proportion - margin_of_error
ci_upper = sample_proportion + margin_of_error

print(f"Sample Proportion: {sample_proportion:.4f}")
print(f"95% Confidence Interval for Proportion: ({ci_lower:.4f}, {ci_upper:.4f})")
```

### Output:
```
Sample Proportion: 0.7500
95% Confidence Interval for Proportion: (0.7056, 0.7944)
```

### Key Differences Between Point Estimate and Interval Estimate:
- **Point Estimate**:
  - Single value estimate of the parameter.
  - E.g., sample mean (\(\bar{x}\)) is a point estimate of the population mean (\(\mu\)).
  - Provides no information about the variability or uncertainty.
  
- **Interval Estimate**:
  - A range of values within which the parameter is likely to fall.
  - E.g., confidence interval for the population mean.
  - Takes into account variability and provides a level of confidence in the estimate.

### Summary:
- **Point estimate**: A single best guess of a population parameter (e.g., sample mean).
- **Interval estimate**: A range (confidence interval) around the point estimate, which provides a level of confidence that the population parameter lies within that range.

````python
#Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
import matplotlib.pyplot as plt
import numpy as np

def bernoulli_pdf(p, x_values=[0, 1]):
    """
    Function to calculate the Probability Distribution Function (PDF)
    of a Bernoulli distribution.

    :param p: Probability of success (0 <= p <= 1)
    :param x_values: Possible outcomes (default is [0, 1])
    :return: A dictionary with outcomes as keys and their probabilities as values
    """
    if not 0 <= p <= 1:
        raise ValueError("p must be between 0 and 1 inclusive.")

    # PDF for Bernoulli: P(X = 1) = p, P(X = 0) = 1 - p
    pdf = {0: 1 - p, 1: p}

    # Output PDF for given values of x (default [0, 1])
    result = {x: pdf[x] for x in x_values}

    return result

# Example usage
p = 0.7  # Probability of success
pdf_values = bernoulli_pdf(p)

print(f"PDF for Bernoulli Distribution with p = {p}:")
print(pdf_values)

# Plotting the PDF
x = [0, 1]
y = [pdf_values[0], pdf_values[1]]

plt.bar(x, y, width=0.1, color='skyblue', edgecolor='black')
plt.title(f'Bernoulli Distribution PDF (p = {p})')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks([0, 1], ['0 (Failure)', '1 (Success)'])
plt.ylim(0, 1)
plt.show()
````

````python
#Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(a, b):
    """
    Function to calculate the Cumulative Distribution Function (CDF)
    of a discrete uniform distribution with range [a, b].

    :param a: The lower bound of the distribution
    :param b: The upper bound of the distribution
    :return: x_values (possible outcomes) and y_values (CDF values)
    """
    # Generate the possible outcomes (x values)
    x_values = np.arange(a, b + 1)

    # Calculate the CDF values for each x
    y_values = (x_values - a + 1) / (b - a + 1)

    return x_values, y_values

# Parameters for the discrete uniform distribution
a = 1  # Lower bound
b = 6  # Upper bound

# Get CDF values
x_values, y_values = discrete_uniform_cdf(a, b)

# Plotting the CDF
plt.figure(figsize=(8, 6))
plt.step(x_values, y_values, where='post', color='skyblue', linewidth=2)

# Adding labels and title
plt.title(f'Cumulative Distribution Function (CDF) of Discrete Uniform Distribution\nRange: [{a}, {b}]', fontsize=14)
plt.xlabel('Outcome', fontsize=12)
plt.ylabel('CDF', fontsize=12)
plt.xticks(np.arange(a, b + 1))  # Ensure all integer values from a to b are labeled
plt.grid(True)

# Show the plot
plt.show()

````

````python
#Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Function to calculate the Z-scores for a dataset.

    :param data: List or numpy array of data points.
    :return: Z-scores of the dataset.
    """
    mean = np.mean(data)  # Calculate the mean of the dataset
    std_dev = np.std(data)  # Calculate the standard deviation of the dataset

    # Calculate the Z-scores for each data point
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(loc=0, scale=1, size=1000)  # Generate a normal distribution of 1000 samples

# Calculate the Z-scores
z_scores = calculate_z_scores(data)

# Plotting the Z-scores
plt.figure(figsize=(8, 6))
plt.hist(z_scores, bins=30, alpha=0.7, color='skyblue', edgecolor='black', density=True)

# Adding labels and title
plt.title('Histogram of Z-scores', fontsize=14)
plt.xlabel('Z-score', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
````

````python
#Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

# Function to implement Central Limit Theorem
def central_limit_theorem(pop_dist, sample_size, num_samples):
    sample_means = []

    # Generate samples and calculate the sample means
    for _ in range(num_samples):
        sample = np.random.choice(pop_dist, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))

    return np.array(sample_means)

# Parameters
population_size = 100000  # Size of the population from which we sample
lambda_exp = 1            # Rate parameter for the exponential distribution
sample_size = 50          # Sample size (number of data points per sample)
num_samples = 1000        # Number of samples to take

# Generate a non-normal population (exponential distribution)
population = np.random.exponential(scale=1/lambda_exp, size=population_size)

# Apply Central Limit Theorem
sample_means = central_limit_theorem(population, sample_size, num_samples)

# Plotting the original population distribution (Exponential)
plt.figure(figsize=(12, 6))

# Plot the population distribution (Exponential)
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Exponential Distribution (Population)', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Plot the distribution of sample means (CLT)
plt.subplot(1, 2, 2)
plt
````

````python
#Generate random variables and calculate their corresponding probabilities using the binomial distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Function to generate random variables and calculate their probabilities using the binomial distribution
def binomial_distribution_probabilities(n, p, size):
    # Generate random variables (number of successes) from the binomial distribution
    random_variables = np.random.binomial(n, p, size)

    # Calculate the probabilities for each value of k (number of successes)
    probabilities = binom.pmf(random_variables, n, p)

    return random_variables, probabilities

# Parameters for the binomial distribution
n = 10   # Number of trials
p = 0.5  # Probability of success in each trial
size = 1000  # Number of random variables to generate

# Generate random variables and calculate their probabilities
random_variables, probabilities = binomial_distribution_probabilities(n, p, size)

# Plotting the histogram of the random variables and the calculated probabilities
plt.figure(figsize=(10, 6))

# Plot histogram of random variables
plt.hist(random_variables, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Random Variables')

# Plot the probability mass function (PMF) of the binomial distribution
x = np.arange(0, n+1)
pmf_values = binom.pmf(x, n, p)
plt.plot(x, pmf_values, 'ro', label='PMF (Theoretical)', markersize=10)

# Adding labels and title
plt.title(f'Binomial Distribution: n = {n}, p = {p}', fontsize=14)
plt.xlabel('Number of Successes (k)', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
````

````python
#Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to calculate the Z-score and compare it to a standard normal distribution
def calculate_z_score(x, mean, std_dev):
    # Calculate Z-score
    z = (x - mean) / std_dev
    return z

# Function to plot the standard normal distribution and the Z-score
def plot_z_score(z):
    # Generate values for the standard normal distribution
    x_values = np.linspace(-4, 4, 1000)
    y_values = stats.norm.pdf(x_values, loc=0, scale=1)

    # Plot the standard normal distribution
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values, label='Standard Normal Distribution', color='blue')

    # Mark the Z-score on the plot
    plt.axvline(x=z, color='red', linestyle='--', label=f'Z-score = {z:.2f}')
    plt.fill_between(x_values, y_values, where=(x_values <= z), color='lightgray', alpha=0.5)

    # Add labels and title
    plt.title(f'Standard Normal Distribution with Z-score = {z:.2f}', fontsize=14)
    plt.xlabel('Z', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()

# Parameters
mean = 50              # Mean of the distribution
std_dev = 10           # Standard deviation of the distribution
data_point = 60        # Data point for which we are calculating the Z-score

# Calculate the Z-score for the given data point
z_score = calculate_z_score(data_point, mean, std_dev)

# Output the Z-score
print(f"Z-score for the data point {data_point}: {z_score:.2f}")

# Plot the standard normal distribution and the Z-score
plot_z_score(z_score)
````

````python
#Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
import scipy.stats as stats

# Function to perform Z-test for hypothesis testing
def z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Calculate sample mean and sample size
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Calculate the Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))  # Two-tailed test

    # Output the results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z:.2f}")
    print(f"P-value: {p_value:.4f}")

    # Make a decision based on p-value
    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")

# Example dataset (sample data)
sample_data = np.array([78, 80, 85, 72, 76, 74, 81, 79, 77, 82])

# Known population mean and standard deviation
population_mean = 75  # Null hypothesis: The population mean is 75
population_std = 5    # Assume known population standard deviation

# Perform Z-test with 95% confidence (alpha = 0.05)
z_test(sample_data, population_mean, population_std, alpha=0.05)
````

````python
#Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

# Function to calculate the confidence interval
def confidence_interval(data, confidence=0.95):
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimator)
    n = len(data)  # Sample size

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(n)

    # Determine the critical value for the confidence level
    confidence_level = confidence
    alpha = 1 - confidence_level
    critical_value = stats.t.ppf(1 - alpha/2, df=n-1)  # t-distribution critical value

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, margin_of_error

# Example dataset (sample data)
data = np.array([78, 80, 85, 72, 76, 74, 81, 79, 77, 82])

# Calculate the confidence interval
lower_bound, upper_bound, sample_mean, margin_of_error = confidence_interval(data, confidence=0.95)

# Output the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
````

````python
#Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats

# Function to calculate the confidence interval
def confidence_interval(data, confidence=0.95):
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimator)
    n = len(data)  # Sample size

    # Calculate the standard error
    standard_error = sample_std / np.sqrt(n)

    # Determine the critical value for the confidence level
    confidence_level = confidence
    alpha = 1 - confidence_level
    critical_value = stats.t.ppf(1 - alpha/2, df=n-1)  # t-distribution critical value

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, margin_of_error

# Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
population_mean = 50
population_std = 10
sample_size = 1000

# Generate a random sample of data from a normal distribution
data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Calculate the confidence interval
lower_bound, upper_bound, sample_mean, margin_of_error = confidence_interval(data, confidence=0.95)

# Output the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
````

````python
#Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to calculate and visualize the PDF of a normal distribution
def plot_normal_pdf(mean, std_dev, num_points=1000):
    # Generate a range of x values from mean - 4*std_dev to mean + 4*std_dev
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, num_points)

    # Calculate the PDF for these x values
    pdf = stats.norm.pdf(x, mean, std_dev)

    # Plot the PDF
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf, label=f'Normal Distribution\n$\mu$={mean}, $\sigma$={std_dev}', color='blue')
    plt.fill_between(x, pdf, alpha=0.2, color='lightblue')

    # Add labels and title
    plt.title('Probability Density Function of a Normal Distribution', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('Probability Density', fontsize=12)
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

# Parameters for the normal distribution
mean = 0        # Mean of the distribution
std_dev = 1     # Standard deviation of the distribution

# Call the function to calculate and plot the PDF
plot_normal_pdf(mean, std_dev)
````

````python
#Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to calculate and plot the CDF of a Poisson distribution
def plot_poisson_cdf(lam, x_max=20):
    # Generate x values from 0 to x_max
    x_values = np.arange(0, x_max+1)

    # Calculate the CDF for each x value
    cdf_values = stats.poisson.cdf(x_values, lam)

    # Plot the CDF
    plt.figure(figsize=(8, 6))
    plt.step(x_values, cdf_values, where="post", label=f'Poisson CDF ($\\lambda$={lam})', color='blue', linewidth=2)

    # Add labels and title
    plt.title('Cumulative Distribution Function of a Poisson Distribution', fontsize=14)
    plt.xlabel('Number of Events (x)', fontsize=12)
    plt.ylabel('Cumulative Probability', fontsize=12)
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()

# Parameters for the Poisson distribution
lambda_param = 4  # Average number of events in a fixed interval

# Calculate and plot the CDF for the Poisson distribution with lambda=4
plot_poisson_cdf(lambda_param)
````

````python
# Simulate a random variable using a continuous uniform distribution and calculate its expected value
import numpy as np

# Function to simulate a continuous uniform distribution and calculate its expected value
def simulate_uniform_distribution(a, b, sample_size=10000):
    # Simulate random variables from a continuous uniform distribution
    random_variables = np.random.uniform(a, b, sample_size)

    # Calculate the theoretical expected value
    expected_value_theoretical = (a + b) / 2

    # Calculate the sample mean (empirical expected value)
    expected_value_empirical = np.mean(random_variables)

    return expected_value_theoretical, expected_value_empirical

# Parameters for the continuous uniform distribution
a = 2  # Lower bound
b = 8  # Upper bound

# Simulate and calculate expected value
theoretical, empirical = simulate_uniform_distribution(a, b)

# Output the results
print(f"Theoretical Expected Value: {theoretical}")
print(f"Empirical Expected Value (Sample Mean): {empirical}")
````

````python
#Write a Python program to compare the standard deviations of two datasets and visualize the difference
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate and compare standard deviations of two datasets
def compare_standard_deviations(data1, data2):
    # Calculate standard deviations
    std_dev1 = np.std(data1)
    std_dev2 = np.std(data2)

    # Print the standard deviations
    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

    # Create side-by-side box plots to visualize the difference
    plt.figure(figsize=(10, 6))
    plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'), medianprops=dict(color='red'))

    # Add title and labels
    plt.title('Comparison of Standard Deviations of Two Datasets', fontsize=14)
    plt.ylabel('Values', fontsize=12)
    plt.show()

# Generate two random datasets with different standard deviations
np.random.seed(42)  # For reproducibility
data1 = np.random.normal(loc=50, scale=10, size=1000)  # Mean = 50, Std dev = 10
data2 = np.random.normal(loc=50, scale=20, size=1000)  # Mean = 50, Std dev = 20

# Compare the standard deviations and visualize the difference
compare_standard_deviations(data1, data2)
# Create side-by-side histograms to visualize the difference
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=30, alpha=0.7, label='Dataset 1', color='blue')
plt.hist(data2, bins=30, alpha=0.7, label='Dataset 2', color='orange')

# Add title and labels
plt.title('Histogram Comparison of Two Datasets', fontsize=14)
plt.xlabel('Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()

plt.show()

````

````python
#Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np

# Function to calculate the Range and Interquartile Range (IQR)
def calculate_range_and_iqr(data):
    # Calculate the Range
    data_range = np.max(data) - np.min(data)

    # Calculate the Interquartile Range (IQR)
    Q1 = np.percentile(data, 25)  # 25th percentile
    Q3 = np.percentile(data, 75)  # 75th percentile
    IQR = Q3 - Q1

    return data_range, IQR

# Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean = 50, Std dev = 10, Sample size = 1000

# Calculate the Range and IQR
data_range, iqr = calculate_range_and_iqr(data)

# Output the results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")
import matplotlib.pyplot as plt

# Create a box plot to visualize the distribution and IQR
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'), medianprops=dict(color='red'))
plt.title('Boxplot of the Normal Distribution Dataset', fontsize=14)
plt.xlabel('Values', fontsize=12)
plt.show()
````

````python
#Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

# Function to apply Z-score normalization
def z_score_normalization(data):
    # Calculate the mean and standard deviation of the data
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply the Z-score formula to normalize the data
    normalized_data = (data - mean) / std_dev

    return normalized_data, mean, std_dev

# Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean = 50, Std dev = 10, Sample size = 1000

# Apply Z-score normalization
normalized_data, original_mean, original_std_dev = z_score_normalization(data)

# Output the original mean and standard deviation
print(f"Original Mean: {original_mean:.2f}")
print(f"Original Standard Deviation: {original_std_dev:.2f}")

# Visualize the original and normalized datasets
plt.figure(figsize=(12, 6))

# Plot the original dataset
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue', label='Original Data')
plt.axvline(original_mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {original_mean:.2f}')
plt.title('Original Dataset Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

# Plot the normalized dataset
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, alpha=0.7, color='green', label='Normalized Data')
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean: 0')
plt.title('Normalized Dataset Distribution (Z-score)')
plt.xlabel('Z-scores')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
````

````python
#Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution
import numpy as np
import scipy.stats as stats

# Function to calculate skewness and kurtosis of a dataset
def calculate_skewness_and_kurtosis(data):
    # Calculate skewness
    skewness = stats.skew(data)

    # Calculate kurtosis (excess kurtosis, so normal distribution will return 0)
    kurtosis = stats.kurtosis(data)

    return skewness, kurtosis

# Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Std dev = 1, Sample size = 1000

# Calculate skewness and kurtosis
skew, kurt = calculate_skewness_and_kurtosis(data)

# Output the results
print(f"Skewness of the dataset: {skew:.2f}")
print(f"Kurtosis (Excess) of the dataset: {kurt:.2f}")
import matplotlib.pyplot as plt

# Plot histogram of the dataset
plt.hist(data, bins=30, alpha=0.7, color='green', density=True)
plt.title('Normal Distribution (Skewness & Kurtosis)')
plt.xlabel('Values')
plt.ylabel('Density')
plt.show()

````
